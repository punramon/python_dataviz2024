* find the first witness lives at the last house on "Northwestern Dr"

```python
# หาบ้านหลังสุดท้ายบนถนน Northwestern Dr
last_house_number = person[person.address_street_name == 'Northwestern Dr']['address_number'].max()

# หา person ที่อยู่บ้านหลังสุดท้ายบนถนน Northwestern Dr
witness1 = person[(person.address_street_name == 'Northwestern Dr') & (person.address_number == last_house_number)]
witness1
```
