```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name'); #ใส่ชื่อcolที่อยากให้แยกสี--แยกสีตาม Name
```

## ปรับ parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#FFD700', '#C0C0C0', '#CD7F32')) #html color (RGB-->str) เลขฐาน16
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0), (0,0.7,0), (0.6,0,1))) #rgb value (r,g,b) | r,g,b -> [0,1] 1สว่างสุด
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r', 'g', 'b')) #ตัวย่อของสี R G B W C M Y K
```

## ปรับ parameter: colormap
* สีที่เขาคิดมาแล้ว
[colormap](https://matplotlib.org/stable/users/explain/colors/colormaps.html)

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='Accent')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy() #เลือกทุกแถวและทุกคอลลัมยกเว้นคอลัมสุดท้าย
np_data
```

# Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure  (ขนาด)

plt.imshow(np_data[:50,:],interpolation='nearest') #show รูป
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='plasma')
```

# การใช้ subplot

* สีเข้มเลขน้อย สีสว่างเลขมาก
* เอากราฟหลายกราฟมารวมกัน
* cmap ไปเลือกสี
* alpha ความจางสี 1 เข้มสุด

```
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest',cmap='magma',alpha = 0.5 )
```
