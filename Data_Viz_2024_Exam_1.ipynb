### 1. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os

path = '/content/drive/MyDrive/Examdv' #set path --> folder and then join path เพื่อให้โค้ดrunได้ทุกที่

tidydata_file_path = os.path.join(path,'test-TidyData_2024.xlsx')
print(tidydata_file_path)
```

```python
tidydata = pd.read_excel(tidydata_file_path)
tidydata
```

```python
# ตัด column '@dropdown' ออก
tidydata = tidydata.drop('@dropdown', axis=1)
tidydata
```

```python
# prompt: ให้แบ่งค่าในcolumn month โดยใช้ _ แบ่ง แล้วเอาค่าตัวเลขที่ถูกแบ่งสร้างเป็น column Year ลบ month เดิมออก

# แบ่งค่าใน column 'month' และสร้าง column 'Year'
tidydata[['Month', 'Year']] = tidydata['month'].str.split('_',expand=True)

# ลบ column 'month' เดิม
tidydata = tidydata.drop('month', axis=1)

# แสดง DataFrame ที่อัปเดตแล้ว
tidydata
```

```python
# prompt: ให้หาความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

# คำนวณค่าเฉลี่ย Chlo_a สำหรับแต่ละ station ในปี 2018
chlo_a_2018 = tidydata[tidydata['Year'] == '2018'].groupby('station')['Chlo_a'].mean()

# คำนวณค่าเฉลี่ย Chlo_a สำหรับแต่ละ station ในปี 2019
chlo_a_2019 = tidydata[tidydata['Year'] == '2019'].groupby('station')['Chlo_a'].mean()

# คำนวณความแตกต่างของค่าเฉลี่ยระหว่างปี 2018 และ 2019
difference = chlo_a_2019 - chlo_a_2018

# แสดงผลลัพธ์
print("ค่าเฉลี่ย Chlo_a ทุกๆ station ระหว่างปี 2018 และ 2019:")
print(difference)
```

```python
# prompt: รวมตาราง chlo_a_2018 chlo_a_2019 และ difference
# รวมตาราง chlo_a_2018, chlo_a_2019 และ difference
result = pd.concat([chlo_a_2018, chlo_a_2019, difference], axis=1)
result.columns = ['Chlo_a_2018', 'Chlo_a_2019', 'Difference']
result
```

### 2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
# prompt: สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

# สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน
chlo_b_mean = tidydata.groupby(['Month', 'station'])['Chlo_b'].mean().reset_index()
chlo_b_mean
```

```python
# แปลงรูปแบบเป็น wide format
chlo_b_mean_pivot = chlo_b_mean.pivot(index=['station'], columns='Month', values='Chlo_b')
chlo_b_mean_pivot
```

```python
# แสดงตาราง
from google.colab import data_table
data_table.DataTable(chlo_b_mean_pivot, include_index=True, num_rows_per_page=6)
```
