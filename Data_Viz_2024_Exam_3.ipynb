เมื่อความคิดเห็นได้รับไลค์มากขึ้น ก็มีแนวโน้มที่จะมีการตอบกลับมากขึ้นด้วย

```python
# คัดเลือก 3 โพสต์ที่มีจำนวนไลค์สูงสุด
top_liked_posts = docter.nlargest(3, 'ยอดไลค์')

# แสดงผลลัพธ์: ชื่อผู้โพสต์, เนื้อหาความคิดเห็น, ยอดไลค์ และจำนวนการตอบกลับ
print("3 เรื่องที่คนสนใจมากที่สุด:")
for i, row in top_liked_posts.iterrows():
    print(f"\nเรื่องที่ {i + 1}:")
    print(f"ชื่อผู้โพสต์: {row['name']}")
    print(f"เนื้อหา: {row['comments']}")
    print(f"ยอดไลค์: {row['ยอดไลค์']}")
    print(f"จำนวนการตอบกลับ: {row['จำนวนการตอบกลับ']}")
```

```python
# กำหนดคำที่เกี่ยวข้องกับการเสียชีวิต
death_keywords = ['ตาย','เสียชีวิต', 'จากไป', 'ไม่รอด', 'สิ้นใจ']

# กำหนดคำที่เกี่ยวข้องกับชนิดของมะเร็ง
cancer_keywords = {
    'มะเร็งปอด': 0,
    'มะเร็งเต้านม': 0,
    'มะเร็งตับ': 0,
    'มะเร็งลำไส้': 0,
    'มะเร็งกระดูก': 0,
    'มะเร็งปากมดลูก': 0,
    'มะเร็งต่อมลูกหมาก': 0,
    'มะเร็งผิวหนัง': 0,
    'มะเร็งกระเพาะอาหาร': 0,
    'มะเร็งเม็ดเลือดขาว': 0,
    'มะเร็งสมอง': 0,
    'มะเร็งรังไข่': 0,
    'มะเร็งต่อมไทรอยด์': 0,
    'มะเร็งต่อมน้ำเหลือง': 0,
    'มะเร็งกระเพาะปัสสาวะ': 0,
    'มะเร็งช่องปาก': 0,
    'มะเร็งหลอดอาหาร': 0
}

# ค้นหาและนับการกล่าวถึงมะเร็งชนิดต่างๆ ที่มีการกล่าวถึงการเสียชีวิต
for comment in docter['comments']:
    if any(keyword in comment for keyword in death_keywords):
        for cancer in cancer_keywords:
            if cancer in comment:
                cancer_keywords[cancer] += 1

# จัดเรียงข้อมูลตามจำนวนการกล่าวถึงจากมากไปน้อย
sorted_cancer_deaths = sorted(cancer_keywords.items(), key=lambda item: item[1], reverse=True)

# แสดงผลลัพธ์ 10 อันดับแรก
top_cancer_deaths = sorted_cancer_deaths[:17]
print("ชนิดของมะเร็งที่มีการกล่าวถึงการเสียชีวิตมากที่สุด:")
for cancer, count in top_cancer_deaths:
    print(f"{cancer}: {count} ครั้ง")
```

```python
# กำหนดคำที่เกี่ยวข้องกับการเสียชีวิต
death_keywords = ['ตาย','เสียชีวิต', 'จากไป', 'ไม่รอด', 'สิ้นใจ', 'ล้มตาย']

# สร้างคอลัมน์ใหม่ที่บ่งบอกว่ามีการกล่าวถึงการเสียชีวิตหรือไม่
docter['เสียชีวิต'] = docter['comments'].apply(lambda x: any(keyword in x for keyword in death_keywords))

# กรองข้อมูลเพื่อแสดงเฉพาะโพสต์ที่มีการกล่าวถึงการเสียชีวิต
death_posts = docter[docter['เสียชีวิต'] == True]

# นับจำนวนโพสต์ที่มีการกล่าวถึงการเสียชีวิต
death_posts_count = len(death_posts)

# แสดงผลลัพธ์จำนวนโพสต์
print(f"จำนวนโพสต์ที่กล่าวถึงการเสียชีวิต: {death_posts_count}")

# แสดงโพสต์ที่กล่าวถึงการเสียชีวิต
print(death_posts[['name', 'comments', 'ยอดไลค์', 'จำนวนการตอบกลับ']])
```
